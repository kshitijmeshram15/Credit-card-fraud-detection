```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_csv('creditcard.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
#looking for missing value
df.isnull().sum()
```

```python
sns.countplot(x='Class',data=df)
```

```python
#It is a highly imbalanced dataset.
df['Class'].value_counts()
```

```python
#separating the data 
legit=df[df['Class']==0]
```

```python
legit
```

```python
fraud=df[df['Class']==1]
```

```python
fraud
```

```python
#analyzing the amount column for legit and fraud data
legit['Amount'].describe()
```

```python
fraud['Amount'].describe()
```

```python
#camparing value of both transaction
df.groupby('Class').mean()
```

```python
#Now dealing with unbalanced data set using under-sampling(removing samples from majority class)so that we get a uniform distribution of data.
legit_sample=legit.sample(n=492)
```

```python
#now concatenating two dataframes.
new_df=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_df.head()
```

```python
new_df['Class'].value_counts()
```

```python
new_df.groupby('Class').mean()
```

```python
#performing train-test-split
x=new_df.drop('Class',axis=1)
y=new_df['Class']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=42)
```

```python
#Model training
from sklearn.linear_model import LogisticRegression
```

```python
log_model=LogisticRegression()
```

```python
log_model.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
#accuracy on training data
train_predictions=log_model.predict(x_train)
```

```python
train_accuracy=accuracy_score(y_train,train_predictions)
```

```python
train_accuracy
```

```python
#Predictions on test data and evaluations
test_predictions=log_model.predict(x_test)
```

```python
test_accuracy=accuracy_score(y_test,test_predictions)
```

```python
test_accuracy
```

```python
from sklearn.metrics import classification_report,confusion_matrix,plot_roc_curve
```

```python
print(classification_report(y_test,test_predictions))
```

```python
print(confusion_matrix(y_test,test_predictions))
```

```python
plot_roc_curve(log_model,x_test,y_test)
```
